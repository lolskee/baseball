````python
from selenium import webdriver
````

````python
driver = webdriver.Firefox()
````

````python
driver.get(pitcher_id_url_1)
````

````python
page1_source = driver.page_source
````

````python
page1_soup = BeautifulSoup(page1_source)
````

````python
header = ('"Player", "Team", "Player ID", "Position", "At Bats",'
    '"Runs", "Hits", "Doubles", "Triples", "Home Runs", "RBIs", "Walks", "Strikeouts", "Stolen Bases",'
    '"Caught Stealing", "Batting Average", "On Base Percentage", "Slugging Percentage", "OPS"')
with open('mlb_data/splits/batters_against_LHP.csv', 'w') as lefties_file:
    lefties_file.write(header)
    for table in page1_soup.findAll('table'):
        for row in table.findAll('tr'):
            row_data = '\n'
    #         for a in row.findAll('a'):
    #             row_data += '"' + a.text + '",'
    #             row_data += '"' + a['href'].split('=')[1] + '",'
            tds = row.findAll('td')
            for i, td in enumerate(tds):
                if i not in [0, 3, 6]:
                    row_data += '"' + td.text + '"'
                    if i != len(tds) - 1:
                        row_data += ','
            lefties_file.write(row_data)
            print(row_data)
            print()
````
