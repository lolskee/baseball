```python
%matplotlib notebook
import matplotlib.pyplot as plt
from urllib.request import urlopen
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
batters_url = 'http://www.rotowire.com/baseball/player_ex_stats.htm'
```

```python
req = urlopen(batters_url)
```

```python
html = req.read()
```

```python
soup = BeautifulSoup(html)
```

```python
soup
```

```python
for table in soup.findAll('table', {'class': 'tablesorter'}):
    for ri, row in enumerate(table.findAll('tr')):
        if ri > 0:
            row_data = '\n'
        else:
            row_data = str()
        spans = row.findAll('span')
        batter_header = str()
        for i, th in enumerate(spans):
            batter_header += th['data-original-title']
            if i != len(spans) - 1:
                batter_header += ','
#         batter_header = 'Last Name, First Name, ' + batter_header
        with open('mlb_data/player_stats/batters.csv', 'a') as batters_file:
            print(batter_header.strip('\n'))
            batters_file.write(batter_header)
            tds = row.findAll('td')
            for i, td in enumerate(tds):
                row_data += td.text
                if i != len(tds) - 1:
                    row_data += ','
            print(row_data.strip('\r'))
            batters_file.write(row_data)
```
