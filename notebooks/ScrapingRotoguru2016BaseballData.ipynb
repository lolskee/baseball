```python
def get_mlb_data_from_date(month, day, fp='mlb_data/DK_2016/'):
    pitchers_header = 'Pitchers, Points, Salary, Team, Opp., Score, Stats Summary'
    hitters_header = 'Hitters, Points, Salary, Team, Opp., Score, Stats Summary'
    day = str(day) if len(str(day)) > 1 else '0' + str(day)
    month_plus_day = str(month) + day
    url = 'http://rotoguru1.com/cgi-bin/byday.pl?date={}&game=dk'.format(month_plus_day)
    req = urlopen(url)
    html = req.read()
    soup = BeautifulSoup(html)
    with open(fp + month_plus_day + '_pitchers.csv', 'w') as pitchers_file:
        with open(fp + month_plus_day + '_hitters.csv', 'w') as hitters_file:
            for table in soup.findAll('table', {"cellspacing": "5"}):
                rows = table.findAll('tr')
                file, player_type = pitchers_file, 'pitchers'
                for i, row in enumerate(rows):
                    row_data = '\n'
                    for j, elem in enumerate(row.findAll()):
                        row_data += elem.text
                        if j != len(row.findAll()) - 1:
                            row_data += ','
                    if 'Hitters' in row_data and i > 1 and player_type != 'hitters':
                        file = hitters_file
                        player_type = 'hitters'
                        file.write(hitters_header)
                    else:
                        if 'Hitters' in row_data:
                            pass
                        elif i == 1:
                            file.write(pitchers_header)
                        elif i > 1 and i < len(rows) - 1:
                            file.write(row_data)
```

```python
get_mlb_data_from_date(4, 3)
```

# Serializing the Process

```python
import datetime as dt
```

```python
from calendar import monthrange
```

```python
monthrange(2016, 4)
```

```python
month_day_tups = list()
```

```python
months = [4, 5, 6, 7, 8, 9]
```

```python
for month in months:
    days_in_month = monthrange(2016, month)[1]
    for i in range(days_in_month):
        month_day_tups.append((month, i + 1))
```

```python
month_day_tups.append((10, 1))
```

```python
month_day_tups.append((10, 2))
```

```python
month_day_tups = month_day_tups[2:]
```

```python
for month_day_tup in month_day_tups:
    month, day = month_day_tup
    get_mlb_data_from_date(month, day)
```
