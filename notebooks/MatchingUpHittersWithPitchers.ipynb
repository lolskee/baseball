```python
def merge_data(day, fp='mlb_data/DK_2016/'):
    hitters = pd.read_csv('mlb_data/DK_2016/{}_hitters.csv'.format(day), 
                      header=None, names=hitter_col_names, 
                      skiprows=[0])
    hitters.reset_index(inplace=True)
    hitters.drop('index', axis=1, inplace=True)
    hitters = hitters[hitters['Player Starting'].str.contains('\^')]
    hitters.dropna(inplace=True)
    hitters['Salary'] = hitters.apply(lambda row: 
        float(row['Salary (Thousands Component)'].strip('$'))
        + float(row['Salary (Ones Component)'])/1e3, axis=1)
    hitters = hitters[['Player', 'Batting Order', 'Team', 'Opponent', 'Salary', 'DK Points']]
    hitters['Batter Home/Away'] = hitters.Opponent.apply(lambda x: 1 if 'v' in x else 0)
    hitters.columns = [
        'Batter', 
        'Batter Batting Order', 
        'Team',
        'Opponent',
        'Batter Salary',
        'Batter DK Points', 
        'Batter Home/Away'
    ]
    hitters.Opponent = hitters.Opponent.apply(lambda x: x[2:].upper())
    hitters['Date'] = [day]*hitters.shape[0]
    pitchers = pd.read_csv('mlb_data/DK_2016/{}_pitchers.csv'.format(day), 
                       header=None, skiprows=[0], names=pitching_col_names)
    if day >= '624':
        pitchers = pd.read_csv('mlb_data/DK_2016/{}_pitchers.csv'.format(day), 
                   header=None, skiprows=[0], names=pitching_col_names + ['Win/Loss'])
        pitchers.drop('Win/Loss', axis=1, inplace=True)
    pitchers.drop('Postion', axis=1, inplace=True)
#     print(pitchers)
    pitchers = pitchers[pitchers['Player Starting'].str.contains('\^')]
#     print(pitchers.head(113))
    pitchers.dropna(inplace=True)
    pitchers['Salary'] = pitchers.apply(lambda row: 
                                  float(str(row['Salary (Thousands Component)']).strip('$'))
                                 + float(row['Salary (Ones Component)'])/1e3, 
                                  axis=1)
    pitchers = pitchers[['Player', 'Batting Order', 'Team', 'Opponent', 'Salary', 'DK Points']]
    pitchers.columns = [
        'Pitcher',
        'Pitcher Batting Order',
        'Pitcher Team',
        'Pitcher Opposing Team',
        'Pitcher Salary',
        'Pitcher DK Points'
    ]
    matched_hitters = hitters.merge(pitchers, left_on='Opponent', right_on='Pitcher Team', how='left')
    batter_stats = pd.read_csv('mlb_data/player_stats/batters.csv',
                          header=None, skiprows=[0])
    batter_stats.columns = [
        'Last Name',
        'First Name',
        'Batter Team',
        'Batter Position',
        'Games',
        'At Bats',
        'Pitches Per Plate Appearance',
        'Bunts',
        'Grounded Into Double Play',
        'Intentional Walks',
        'Walks',
        'Strikeouts',
        'Walk Rate',
        'Walk-To-Strikeout Ratio',
        'Contact Rate',
        'Stolen Base Opportunity',
        'Batting Average',
        'Batting Average on Balls In Play'
    ]
    batter_stats['Batter'] = batter_stats['First Name'] + " " + batter_stats['Last Name']
    batter_stats['Bunt Rate'] = batter_stats.Bunts/batter_stats['At Bats']
    batter_stats['Ground Into Double Play Rate'] = batter_stats['Grounded Into Double Play']/batter_stats['At Bats']
    batter_stats['Intentional Walk Rate'] = batter_stats['Intentional Walks']/batter_stats['At Bats']
    batter_stats['Strikeout Rate'] = batter_stats['Strikeouts']/batter_stats['At Bats']
    batter_stats.Batter = batter_stats.Batter.apply(lambda x: str(x)[1:])
    hitters_data = matched_hitters.merge(batter_stats, left_on='Batter', right_on='Batter', how='left')
    hitters_data.columns = ['Batter', 'Batter Batting Order', 'Team', 'Opponent', 'Batter Salary',
       'Batter DK Points', 'Batter Home/Away', 'Date', 'Pitcher',
       'Pitcher Batting Order', 'Pitcher Team', 'Pitcher Opposing Team',
       'Pitcher Salary', 'Pitcher DK Points', 'Last Name', 'First Name',
       'Batter Team', 'Batter Position', 'Batter Games', 'At Bats',
       'Pitches Per Plate Appearance', 'Bunts', 'Grounded Into Double Play',
       'Intentional Walks', 'Walks', 'Strikeouts', 'Walk Rate',
       'Walk-To-Strikeout Ratio', 'Contact Rate', 'Stolen Base Opportunity',
       'Batting Average', 'Batting Average on Balls In Play', 'Bunt Rate',
       'Ground Into Double Play Rate', 'Intentional Walk Rate',
       'Strikeout Rate']
    pitcher_stats = pd.read_csv('mlb_data/player_stats/pitchers.csv',
                          header=None, skiprows=[0])
    pitcher_stats.columns = [
        'Last Name',
        'First Name',
        'Pitcher Team',
        'Games',
        'Innings Pitched',
        'Games Started',
        'Quality Starts',
        'Quality Start Rate',
        'Strikeout-to-Walk Ratio',
        'Strikeouts Per 9 IP',
        'Walks Per 9 IP',
        'Home Runs Allowed Per 9 IP',
        'Pitcher Batting Average On Balls In Play',
        'Strand Rate',
        'Fastball Velocity (MPH)',
        'Earned Run Average',
        'Fielding Independent Pitching'
    ]
    pitcher_stats['Pitcher'] = pitcher_stats['First Name'] + ' ' + pitcher_stats['Last Name']
    pitcher_stats.drop(['Last Name', 'First Name'], axis=1, inplace=True)
    pitcher_stats.Pitcher = pitcher_stats.Pitcher.apply(lambda x: str(x)[1:])
    combined_data = hitters_data.merge(pitcher_stats, on='Pitcher', how='left')
    combined_data.dropna(inplace=True)
    return combined_data
```
